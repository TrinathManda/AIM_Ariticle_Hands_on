```
from ExploriPy import EDA
import pandas as pd
```

```
df = pd.read_csv('Advertising.csv')
```

```
df.drop(columns='Unnamed: 0', inplace=True, axis=1)
```

```
ContinuousFeatures = ['tadio','newspaper','TV']
analysis = EDA( df, debug='YES', filename='index.html',title='EDA for Sales Data',)
```
