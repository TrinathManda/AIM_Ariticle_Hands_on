```python
x = np.arange(-10,10,0.1)
```

```python
def sinh(x):
  return 0.5 * (np.exp(x) - np.exp(-x))

y = [sinh(val) for val in x]

px.line(x=x, y=y)
```

```python
def cosh(x):
  return 0.5 * (np.exp(x) + np.exp(-x))

y = [cosh(val) for val in x]

px.line(x=x, y=y)
```

```python
def tanh(x):
  return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

y = [tanh(val) for val in x]

px.line(x=x, y=y)
```

```python
#Derivative of tanh
def derivative_tanh(x):
  return  1 - np.power(tanh(x), 2)

y = [derivative_tanh(val) for val in x]

px.line(x=x, y=y)
```

```python
#derivative of sigmoid
def dervative_sigmoid(x, w=1, b=0):
  return sigmoid(x, w, b) * (1 - sigmoid(x, w, b))
```

```python
y10 = [sigmoid(val) for val in x]
dy10 = [dervative_sigmoid(val) for val in x]

fig = px.line(x=x, y=y10)
fig.add_scatter(x=x, y=dy10, mode='lines')
fig.show()
```

```python
import ipywidgets as widgets
from ipywidgets import interact, interact_manual
from IPython.display import clear_output

x = np.arange(-10, 10, 0.5)
@interact
def plotSigmoid(slope=(0.1, 50, 0.5)):
  yw0 = [sigmoid(val, slope, 0) for val in x]
  dyw0 = [dervative_sigmoid(val, slope, 0) for val in x]

  clear_output(True)

  fig = px.line(x=x, y=yw0)
  fig.add_scatter(x=x, y=dyw0, mode='lines')
  fig.update_layout(autosize=False, width=1600, height=500)
  fig.show()
```

```python
#ELU
x = np.arange(-10, 10, 0.1)
ALPHA = 1.6732632423543772848170429916717

def elu(x):
  return x if x >= 0 else ALPHA *(np.exp(x) - 1)

y = [elu(val) for val in x]

px.line(x=x, y=y)
```

```python
ALPHA = 1.6732632423543772848170429916717
LAMBDA = 1.0507009873554804934193349852946

def selu(x):
  return LAMBDA * x if x >= 0 else LAMBDA * ALPHA * ( np.exp(x) - 1)

y = [selu(val) for val in x]

px.line(x=x, y=y)
```

```python
def relu(x):
  return x if x >= 0 else 0

y = [relu(val) for val in x]

px.line(x=x, y=y)
```

```python
def derivative_relu(x):
    return 1 if relu(x) >= 0 else 0

y = [derivative_relu(val) for val in x]

px.line(x=x, y=y)

```

```python
x = np.arange(-5, 5, 0.1)
y = [relu(val) for val in x]
dy = [derivative_relu(val) for val in x]
fig = px.line(x=x, y=y)
fig.add_scatter(x=x, y=dy, mode='lines')
fig.show()
```

```python
x = np.arange(-10, 10, 0.1)

ALPHA = np.exp(-2)

def leakyRelu(x):
  return x if x >= 0 else ALPHA * x

y = [leakyRelu(val) for val in x]

px.line(x=x, y=y)
```

```python
def derivative_leakyRelu(x):
  return 1 if x >= 0 else ALPHA

dy = [derivative_leakyRelu(val) for val in x]

px.line(x=x, y=dy)
```
