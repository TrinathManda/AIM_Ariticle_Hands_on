```
import pandas as pd
import lens
```

```
df = pd.read_csv('Advertising.csv')
df.drop(columns='Unnamed: 0', inplace=True, axis=1)

#df.columns
#df.columns=['Preg', 'Plas', 'Pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
```
