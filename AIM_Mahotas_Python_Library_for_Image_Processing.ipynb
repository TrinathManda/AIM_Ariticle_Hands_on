```
import mahotas
from mahotas import *
from pylab import imshow,show
```

```
img = mahotas.imread('Transformers.jpg')
imshow(img)
```

```
img1 = img[:, :, 2]
```

```
mean = img1.mean()
print("Mean Value ", mean) 
imshow(img1)
```

```
#Crop
img2 = img[:, 300:1200] 
imshow(img2)
```

```
img3 = img[:, :, 1]
ecc = mahotas.features.eccentricity(img3)
rnds = mahotas.features.roundness(img3)
print("Eccentricity =", ecc ) 
print("Roundness = ", rnds)
```

```
dilt = mahotas.morph.dilate(img)
imshow(dilt)
```

```
erd = mahotas.morph.erode(img)
imshow(erd)
```

```
img2 = img.max(2)
lmaxim = mahotas.locmax(img2)
imshow(lmaxim)
```

```
rmaxim = mahotas.regmax(img2)
imshow(rmaxim)
```

```
img5 = img[:, :, 0] 
ol = mahotas.overlay(img5)
imshow(ol) 
```

```
import matplotlib.pyplot as plt 

img_hist = mahotas.fullhistogram(img)
plt.hist(img_hist)
```
